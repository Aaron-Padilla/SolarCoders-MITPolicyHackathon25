# Solar Generation within the PJM Energy Grid

```python
# Initial data and dependency imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from itertools import product

data_raw = pd.read_csv(r'./datasets/solar_gen.csv')

data = data_raw.copy(deep = True) # this will be the data to manipulate
```

```python
print(data.columns, '\n\n')

print(data.area.unique())
```

```python
for i in list(data.area.unique()):

    workingData = data[data.area ==i]
    
    plt.plot(workingData.datetime_beginning_ept, workingData.solar_generation_mw)
    plt.title(f'Historical Solar Generation in the PJM grid,\n{i} Area\n{max(workingData.datetime_beginning_ept)[:9]}')
    plt.show()
```
